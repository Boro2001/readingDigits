```python
import pickle 
import numpy as np
import pandas as pd
import os
import matplotlib

infile = open("sgd_clf.pkl",'rb')
sgd_clf = pickle.load(infile)
```

```python
import cv2 as cv
## capturing video from camera
def rescaleFrame(frame, scale=.5):
    dim = (int(frame.shape[1] * scale), int(frame.shape[0]*scale))
    return cv.resize(frame, dim, interpolation=cv.INTER_AREA)

def rescaleFrame(frame, width=28, height=28):
    dim = (int(width), int(height))
    return cv.resize(frame, dim, interpolation=cv.INTER_AREA)


capture = cv.VideoCapture(1)
while True:
    isTrue, frame = capture.read()
    frame = rescaleFrame(frame,28,28)
    frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    #cv.line(frame, (0,0), (frame.shape[1], frame.shape[0]), (255,255,255), thickness=3)
    frame = cv.GaussianBlur(frame,(1,1),cv.BORDER_DEFAULT)
    #frame = cv.Canny(frame, 125,175)
    #frame = cv.dilate(frame, (27,27),iterations=1)
    #frame = frame[50:300, 200:600]
    X = pd.DataFrame(data=frame)
    result = sgd_clf.predict(X.values.ravel().reshape(1,-1))
    os.system('clear')

    cv.putText(frame, str(result), (12,12), cv.FONT_HERSHEY_TRIPLEX,.5,(0,255,0),1)
    cv.imshow('VideoResized', frame)
    if cv.waitKey(20) & 0xFF==ord('d'):
        break


capture.release()
cv.destroyAllWindows()
```
