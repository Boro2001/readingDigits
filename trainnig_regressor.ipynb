```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
mnist.keys()
```

```python
X, y = mnist["data"], mnist["target"].astype(np.uint8)
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
sgd_clf = Pipeline([("scaler", StandardScaler()), ("linear_svc", LinearSVC(C=1, loss='hinge'))])
sgd_clf.fit(X, y)
```

```python
sgd_clf.predict(X[4].reshape(1,-1))
```

```python
import pickle
with open('sgd_clf.pkl', 'wb') as f_conf:
    pickle.dump(sgd_clf, f_conf)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y, sgd_clf.predict(X)))
```
